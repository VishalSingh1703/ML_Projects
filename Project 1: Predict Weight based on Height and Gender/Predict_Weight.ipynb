# Importing Libraries

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# Data Preparation and Analysis

```python
# data = pd.read_csv("weight-height.csv")
# data.to_csv('data.csv')
# data
data = pd.read_csv("weight-height.csv")
data['Weight'] = data['Weight'] - 15
data
```

```python
data.isnull().sum()
```

```python
data.describe()
```

Converting Female and Male to 1s and 0s using LabelEncoder

```python
e = LabelEncoder()
# X = data.iloc[:, :-1].values
# y = data.iloc[:, 2].values
# X[:,0] = e.fit_transform(X[:,0])
data['Gender'] = e.fit_transform(data['Gender'])
data
```

# Model Planning and Building

Splitting the data into training and test sets

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, 2].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Model Build

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

Make Prediction using Test data

```python
lin_pred = lin_reg.predict(X_test)
```

# Run Output

Predict Weight

```python
g = int(input("Enter 1 for female and 0 for male in cm: "))
h = int(input("Enter the height in cm: ")) / 2.54
user_weight = lin_reg.predict([[g, h]]) / 2.2
print(str(round(user_weight[0], 2)) + "kgs")
```
